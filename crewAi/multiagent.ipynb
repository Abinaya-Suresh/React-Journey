```python
import os
os.environ["GEMINI_API_KEY"] = ""
```

```python
from crewai import LLM

llm = LLM(
    model="gemini/gemini-2.0-flash",
    temperature=0.1,
    api_key=os.environ["GEMINI_API_KEY"]
)

print(llm.call("What is CrewAI?"))
```

```python
from crewai import Agent, Task, Crew

research_agent = Agent(
    role="Research Agent",
    goal="Research and find the best way to improve emails",
    backstory="A highly experienced communication expert skilled in professional email writing",
    verbose=True,#detail info or output about execution of a program
    llm=llm
)

writer_agent= Agent(
    role="Writer",
    goal="Write a professional email to a client",
    backstory="A highly experienced communication expert skilled in professional email writing",
    verbose=True,
    llm=llm

)
```

```python
research_task=Task(

    description="Find 3-5 interesting and recent facts about {topic}.",
    expected_output="A bullet list od 3-5 facts",
    agent=research_agent
)

email_task=Task(
    description="Write a 100-words blog post summary about {topic} using using the facts from the research",
    agent=writer_agent,
    expected_output="A blog post summary",
    context=[research_task]
)
```

```python
crew=Crew(
    agents=[research_agent, writer_agent],
    tasks=[research_task, email_task],
    verbose=True
)
```

```python
crew.kickoff(inputs={"topic":"The future of electrical vehicles"})
```
