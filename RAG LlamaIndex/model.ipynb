```python
import os

from dotenv import load_dotenv
load_dotenv()

```

```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader

#documents=SimpleDirectoryReader('RAG Methods list.pdf').load_data()

folder_path = r"C:\Users\abina\Desktop\RAG LlamaIndex"

documents = SimpleDirectoryReader(folder_path).load_data()
```

```python
documents
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding


embedding=HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

index=VectorStoreIndex.from_documents(documents,embed_model=embedding,show_progress=True)
```

```python
index
```

```python
retriever = index.as_retriever()
```

```python
results = retriever.retrieve("What are the data extraction methods")
```
